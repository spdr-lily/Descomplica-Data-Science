# Colocando tokens em uma tabela

```python
import pandas as pd
```

```python
colunas = "Token Lema Letras Formato POS Função Governante Stopword Entidade IOB".split(" ")
```

```python
colunas
```

```python
registros = []
for token in list(doc.sents)[0]:
    registros.append([token.text, token.lemma_, token.is_alpha, token.shape_, token.pos_, token.dep_, token.head, token.is_stop, token.ent_type_, token.ent_iob_])
```

```python
print(registros)
```

```python
tabela = pd.DataFrame(registros, columns=colunas)
```

```python
tabela
```
